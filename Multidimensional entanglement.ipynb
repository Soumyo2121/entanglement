```python
from qiskit import *
#a quantum circuit with 4 qubits(2 for each particle)
multidimensional_entanglement = QuantumCircuit(4)
multidimensional_entanglement.h(0)
multidimensional_entanglement.cx(0,2)
multidimensional_entanglement.h(1)
multidimensional_entanglement.cx(1,3)
multidimensional_entanglement. barrier()
multidimensional_entanglement.h(0)
multidimensional_entanglement.cx(0,2)
multidimensional_entanglement.h(1)
multidimensional_entanglement.cx(1,3)
multidimensional_entanglement.draw('mpl')
```

```python
from qiskit import QuantumCircuit, transpile, Aer, IBMQ
#from qiskit.tools.jupyter import *
from qiskit.visualization import *
from qiskit.providers.aer import QasmSimulator
def execute_circuit_sv(quantum_circuit):
    statevector_simulator = Aer.get_backend('statevector_simulator')
    result = execute(quantum_circuit, statevector_simulator).result()
    statevector_results = result.get_statevector(quantum_circuit)
    circuit_diagram = quantum_circuit.draw('mpl')
    b_sphere = plot_bloch_multivector(statevector_results)
    
    return statevector_results, circuit_diagram, b_sphere
```

```python
result, img, bsphere = execute_circuit_sv(multidimensional_entanglement)
img.savefig('img.png')
bsphere.savefig('bsphere.png')
```

```python
qc = QuantumCircuit(4)
qc.h(0)
qc.cx(0,2)
qc.barrier()
qc.h(1)
qc.cx(1,3)
qc.z(1)
qc.barrier()
qc.cx(0,1)
qc.cx(1,3)
qc.cz(0,3)
qc.draw('mpl')
```

```python
qc1 = QuantumCircuit(4) # 2qubits per particle
qc1.h(0)
qc1.cx(0,1)
qc1.barrier() # entangled the 1st two qubits so that they behave like a single paricle
qc1.h(2)
qc1.cx(2,3)
qc1.barrier()  # entangled the last two qubits so that they behave like a single paricle
qc1.h(0)
qc1.cx(0,2)
qc1.barrier() # entangling the first quantum property
qc1.h(1)
qc1.cx(1,3) # entangling the 2nd quantum property. 
qc1.draw('mpl')
```

```python
result, img2, bsphere2 = execute_circuit_sv(qc1)
img2.savefig('img2.png')
bsphere2.savefig('bsphere2.png')
```
